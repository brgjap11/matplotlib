```python
#importing dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing & reading CSV
city_csv = ("../raw_data/city_data.csv")
ride_csv = ("../raw_data/ride_data.csv")
city_df= pd.read_csv(city_csv, index_col="city")
ride_df = pd.read_csv(ride_csv)
```

```python
#average fare per city
avg_group=ride_df.groupby("city")
avg_fare= avg_group["fare"].mean()
```

```python
# number of rides per city
rides_group = ride_df.groupby("city")
rides_per_city= rides_group["fare"].count()
```

```python
#creating dataframe of two calculations of average fare per city and number of rides per city to join to the city_df
all_calcs= pd.DataFrame({"Average Fare":avg_fare,"Rides Per City":rides_per_city})
```

```python
#merging calcs into the city df
alldata= pd.merge(city_df, all_calcs, left_index=True, right_index=True)
```

```python
alldata_new= alldata.reset_index()
```

```python
#Bringing in Type Data for Bubble chart
onlyurban=alldata_new.loc[alldata_new["Type"]=="Urban",:]
onlyrural=alldata_new.loc[alldata_new["Type"]=="Rural",:]
onlysuburban=alldata_new.loc[alldata_new["Type"]=="Suburban",:]

# urban 
urban_x= onlyurban["Average Fare"]
urban_y= onlyurban["Rides Per City"]
urban_z= onlyurban["driver_count"]
# rural 
rural_x= onlyrural["Average Fare"]
rural_y= onlyrural["Rides Per City"]
rural_z= onlyrural["driver_count"]
#suburban 
suburban_x= onlysuburban["Average Fare"]
suburban_y= onlysuburban["Rides Per City"]
suburban_z= onlysuburban["driver_count"]

plt.title("Pyber Ride Sharing Data")
plt.xlabel("Rides Per City")
plt.ylabel("Drivers Per City")

urban_plt= plt.scatter(urban_y, urban_x, s=urban_z*10, c="gold", linewidth=1.0, alpha=0.65, edgecolors="black", label="Urban")
rural_plt= plt.scatter(rural_y, rural_x, s=rural_z*10, c="lightskyblue",linewidth=1.0,alpha=0.65,  edgecolors="black", label="Rural")
suburban_plt= plt.scatter(suburban_y, suburban_x, s=suburban_z*10, c="lightcoral",linewidth=1.0,alpha=0.65,  edgecolors="black", label="Suburban")

plt.legend(handles=[urban_plt, rural_plt, suburban_plt], loc="best")
plt.show()
```

```python
#% of Total Drivers by City Type

#creating the piechart
colors = ["gold","lightskyblue","lightcoral"]
explode = (0,0,0)
labels= ["Urban","Rural","Suburban"]
rides_by_type=[urban_driver, rural_driver, suburban_driver]

plt.pie(rides_by_type,labels=labels, explode=explode, autopct="%1.1f%%", colors=colors )

# Add Title
plt.title("% of Total Drivers by City Type")

plt.axis("equal")
plt.show()

```
